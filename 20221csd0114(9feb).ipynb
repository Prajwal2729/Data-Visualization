```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
print('Step1')
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,5,20)
ax.plot(x,np.cos(x));
print(x)
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,5,30)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
ax.plot(x,np.tan(x));
print(x)
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,5,30)
ax.plot(x,np.sin(x));
ax.set_title('Plotting')
ax.set_xlabel('Wavelength')
ax.set_ylabel('Amplitude');
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,5,30)
ax.plot(x,np.sin(x),label='sin',color='r',linestyle='dotted')
ax.plot(x,np.cos(x+1),label='sin x+1',color='r',linestyle='solid')
ax.plot(x,np.cos(x),label='cos',color='g',linestyle='dashdot')
ax.set_title('Plotting')
ax.set_xlabel('Wavelength')
ax.set_ylabel('Amplitude');
ax.legend()
plt.show()
```
